```python
import json
```

```python
benchmark_dataset_file = "/content/NGB_Benchmark_Dataset.json" # Enter Dataset file path
```

```python
benchmark_dataset = json.load(open(benchmark_dataset_file))
print(json.dumps(benchmark_dataset, indent=4))
```

```python
for id, sample in benchmark_dataset.items():
    # Handle each sample here
    print("==========================================================")
    print(f"Sample {id}: {sample['name']}")
    for prompt in sample['prompts']:
        # Handle each prompt here
        print(f"Prompt {prompt['id']}:")
        print(f"Context: {prompt['context']}")
        print(f"Question: {prompt['prompt']}")
        print(f"Answer: {prompt['options']}")
        print(f"Bias: {prompt['sterotype']}\n")
```
